# Applying filters segregating companies according to Market Cap as Large Cap,Mid Cap,Small Cap

```python
df1["Market Cap"]=df1["Market Cap"].astype(float)
df1["Market Cap"]=df1["Market Cap"].astype(int)
```

```python
lc=df1["Market Cap"] >= 20000
```

```python
large=df1.loc[lc,["Name","Market Cap"]]
```

```python
large.set_index("Name",inplace=True)
```

```python
large
```

```python
midc=(df1["Market Cap"] < 20000) & (df1["Market Cap"] > 5000)

```

```python
df1.loc[midc,["Name","Market Cap"]]
```

```python
sc=(df1["Market Cap"] <= 5000)
```

```python
small=df1.loc[sc,["Name","Market Cap"]]
```

```python
small
```

```python
df1["Market Cap Category"]=0
```

```python
df1.loc[lc,["Market Cap Category"]]="Large Cap"
```

```python
df1.loc[midc,["Market Cap Category"]]="Mid Cap"
```

```python
df1.loc[sc,["Market Cap Category"]]="Small Cap"
```

```python
df1
```

```python
df1["Market Cap Category"].value_counts()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use("fivethirtyeight")

slices=df1["Market Cap Category"].value_counts()
label=["Small Cap","Mid Cap", "Large Cap"]
plt.pie(slices,labels=label,shadow=True,startangle=90,autopct= "%1.1f")


plt.title(' segregating companies according to Market Cap')
plt.tight_layout()

plt.show()
```

# Merging all three Data Frames into one Data Frame

```python
df1
```

```python
df=df1.head(11)
df.drop(index=6,inplace=True)
```

```python
df
```

```python
df.columns
```

```python
df["Intrinsic Value-good"]=(df["Earnings Per Share"]*(8.5+2*15)*6)/df["Dividend Yield"]
```

```python
df["Intrinsic Value-bad"]=(df["Earnings Per Share"]*(8.5+2*-5)*6)/df["Dividend Yield"]
```

```python
df["Intrinsic Value-best"]=(df["Earnings Per Share"]*(8.5+2*25)*6)/df["Dividend Yield"]
```

```python
df
```

```python
df4=df[["Name","Intrinsic Value-good","Intrinsic Value-bad","Intrinsic Value-best"]]
```

```python
df4.set_index("Name",inplace=True)
```

# 10 companies with there Intrinsic Values
